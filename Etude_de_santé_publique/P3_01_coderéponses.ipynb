<h3> Concaténation des fichiers animaux et végétaux </h3> 

```python
df_animaux['Type de produit'] = 'Animal'
df_vegetaux['Type de produit'] = 'Végétal'
```

```python
df_vegetaux_animaux = df_vegetaux.append(df_animaux)
```

```python
df_vegetaux_animaux
```

```python
df_vegetaux_animaux['Élément'].unique()
```

```python
df_vegetaux_animaux_pivot = df_vegetaux_animaux.pivot_table(index=['Code zone', 'Zone','Code Produit','Produit','Type de produit','Année'],values=['Valeur'],aggfunc = sum,columns=['Élément'])
df_vegetaux_animaux_pivot.columns = ['Aliments pour animaux', 
                               'Autres utilisations (non alimentaire)',
                              'Disponibilité alimentaire (Kcal/personne/jour)',
                              'Disponibilité alimentaire en quantité (kg/personne/an)',
                              'Disponibilité de matière grasse en quantité (g/personne/jour)',
                              'Disponibilité de protéines en quantité (g/personne/jour)',
                              'Disponibilité intérieure',
                              'Exportations - Quantité',
                              'Importations - Quantité',
                              'Nourriture',
                              'Pertes',
                              'Production',
                              'Semences',
                              'Traitement',
                              'Variation de stock']
df_vegetaux_animaux_pivot = df_vegetaux_animaux_pivot.reset_index()
df_vegetaux_animaux_pivot
```

<h3> Sélection des lignes correspondantes </h3>

```python
dispo_int = int(df_vegetaux_animaux_pivot["Disponibilité intérieure"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
production = int(df_vegetaux_animaux_pivot["Production"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
importations = int(df_vegetaux_animaux_pivot["Importations - Quantité"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
exportation = int(df_vegetaux_animaux_pivot["Exportations - Quantité"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
variation_stock = int(df_vegetaux_animaux_pivot["Variation de stock"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
nourriture = int(df_vegetaux_animaux_pivot["Nourriture"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
pertes = int(df_vegetaux_animaux_pivot["Pertes"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")]) 
semences = int(df_vegetaux_animaux_pivot["Semences"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")]) 
aliments_animaux = int(df_vegetaux_animaux_pivot["Aliments pour animaux"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
traitement = int(df_vegetaux_animaux_pivot["Traitement"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")]) 
autres_utilisations = int(df_vegetaux_animaux_pivot["Autres utilisations (non alimentaire)"].loc[(df_vegetaux_animaux_pivot["Zone"] == "France") & (df_vegetaux_animaux_pivot["Produit"]=="Blé")])
```

<h3> Illustrations des équations avec l'exemple du blé en France en 2013 </h3> 

<i> <b> Equation 1 </b> </i>

```python
print("En utilisant l'équation 1 la disponibilité intérieure en France du blé est de " + str(int(dispo_int)) + " milliers de tonnes")
```

<i> <b> Equation 2 </b> </i>

```python
Eq_2= int(production)+ int(importations) + int(variation_stock) - int(exportation)
print("En utilisant l'équation 2 la disponibilité intérieure en France du blé est de " + str(int(Eq_2)) + " milliers de tonnes")
```

<i> <b> Equation 3 </b> </i>

```python
Eq_3= int(nourriture)+int(pertes)+int(semences)+int(aliments_animaux)+int(traitement)+int(autres_utilisations)
print("En utilisant l'équation 3 la disponibilité intérieure en France du blé est de " + str(int(Eq_3)) + " milliers de tonnes")
```
