<h3> Concaténation des fichiers animaux et végétaux </h3> 

```python
df_animaux['Type de produit'] = 'Animal'
df_vegetaux['Type de produit'] = 'Végétal'
```

```python
df_vegetaux_animaux_pivot = df_vegetaux_animaux.pivot_table(index=['Code zone', 'Zone','Code Produit','Produit','Type de produit','Année'],values=['Valeur'],aggfunc = sum,columns=['Élément'])
df_vegetaux_animaux_pivot.columns = ['Aliments pour animaux', 
                               'Autres utilisations (non alimentaire)',
                              'Disponibilité alimentaire (Kcal/personne/jour)',
                              'Disponibilité alimentaire en quantité (kg/personne/an)',
                              'Disponibilité de matière grasse en quantité (g/personne/jour)',
                              'Disponibilité de protéines en quantité (g/personne/jour)',
                              'Disponibilité intérieure',
                              'Exportations - Quantité',
                              'Importations - Quantité',
                              'Nourriture',
                              'Pertes',
                              'Production',
                              'Semences',
                              'Traitement',
                              'Variation de stock']
df_vegetaux_animaux_pivot = df_vegetaux_animaux_pivot.reset_index()
```

```python
nombre_de_personnes=df_population_mondiale[["Zone","Valeur","Unité"]]
nombre_de_personnes
```

```python
Disponibilité_personnes = df_vegetaux_animaux_pivot.merge(nombre_de_personnes,how='inner',on='Zone')
```

<h3> Les 10 pays ayant le plus haut ratio disponibilité alimentaire/habitant en termes de protéines (en kg) par habitant

```python
df_10_pays_plus_dispo_alim_prot_par_hab = pd.read_sql_query(''' SELECT pays,
(SUM (dispo_prot)*365)/1e3 AS "dispo_alim_prot_par_hab"
FROM dispo_alim
GROUP BY dispo_alim.code_pays
ORDER BY dispo_alim_prot_par_hab DESC
LIMIT 10;

''',con)

df_10_pays_plus_dispo_alim_prot_par_hab
```

<h3> Les 10 pays ayant le plus haut ratio disponibilité alimentaire/habitant en termes de kcal par habitant.

```python
df_10_pays_plus_dispo_alim_kcal_par_hab = pd.read_sql_query(''' SELECT pays,
(SUM (dispo_alim_kcal_p_j)*365) AS "dispo_alim_kcal_par_hab"
FROM dispo_alim
GROUP BY dispo_alim.code_pays
ORDER BY dispo_alim_kcal_par_hab DESC
LIMIT 10;

''',con)

df_10_pays_plus_dispo_alim_kcal_par_hab
```

<h3> Pour l'année 2013, les 10 pays ayant le plus faible ratio disponibilité alimentaire/habitant en termes de protéines (en kg) par habitant.

```python
df_10_pays_moins_ratio_dispo_alim_prot_par_hab = pd.read_sql_query(''' SELECT pays,
(SUM (dispo_prot)*365)/1e3 AS "dispo_alim_prot_par_hab"
FROM dispo_alim
GROUP BY dispo_alim.code_pays
ORDER BY dispo_alim_prot_par_hab ASC
LIMIT 10;

''',con)

df_10_pays_moins_ratio_dispo_alim_prot_par_hab
```

<h3> La quantité totale (en kg) de produits perdus par pays en 2013.

```python
df_quantié_totale_kg_perdus_pays = pd.read_sql_query(''' SELECT pays, 
(SUM (pertes)*1e6) AS "pertes_totales_par_pays"
FROM equilibre_prod
GROUP BY equilibre_prod.code_pays
ORDER BY pertes_totales_par_pays DESC;

''',con)

df_quantié_totale_kg_perdus_pays
```

<h3> Les 10 pays pour lesquels la proportion de personnes sous-alimentées est la plus forte

```python
df_10_pays_plus_sous_nutrition = pd.read_sql_query(''' SELECT sous_nutrition.pays, population.population, sous_nutrition.nb_personnes,
((sous_nutrition.nb_personnes*1e6)/(population.population*1000))*100 AS "proportion_sous_nutrition"
FROM sous_nutrition
JOIN population ON sous_nutrition.code_pays = population.code_pays
ORDER BY proportion_sous_nutrition DESC
LIMIT 10;

''',con)

df_10_pays_plus_sous_nutrition
```

<h3> Les 10 produits pour lesquels le ratio Autres utilisations/Disponibilité intérieure est le plus élevé.

```python
df_10_produit_plus_ratio_autres_utilisations_dispo_int = pd.read_sql_query(''' SELECT produit,
(SUM(autres_utilisations))/(SUM(dispo_int)) AS "ratio_autres_utilsation_dispo_int"
FROM equilibre_prod
GROUP BY equilibre_prod.code_produit
ORDER BY ratio_autres_utilsation_dispo_int DESC
LIMIT 10;

''',con)

df_10_produit_plus_ratio_autres_utilisations_dispo_int
```
