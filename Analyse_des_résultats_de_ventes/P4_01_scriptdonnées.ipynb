<h3>Détermination des doublons

```python
df_clients_doublons=df_clients[df_clients.duplicated(keep='first')]
df_clients_doublons
```

```python
df_produits_doublons=df_produits[df_produits.duplicated(keep='first')]
df_produits_doublons
```

```python
df_ventes_doublons=df_ventes_sans_test[df_ventes_sans_test.duplicated(keep='first')]
df_ventes_doublons
```

<h3> Merge des tableaux

```python
df_clients_ventes=df_ventes_sans_test.merge(df_clients,how='inner',on='client_id')
df_clients_ventes
```

```python
df_clients_ventes_produits=df_clients_ventes.merge(df_produits, how='left',on='id_prod')
df_clients_ventes_produits
```

<h3> Remplacement des NA de la colonne price par la moyenne des prix

```python
df_clients_ventes_produits.price.fillna(df_clients_ventes_produits[df_clients_ventes_produits['categ']==0].price.mean(),inplace=True)
```

```python
df_clients_ventes_produits[df_clients_ventes_produits['price'].isnull()==True]
```

<h3> Remplacement des NA de la colonne catego par 0

```python
df_clients_ventes_produits.categ.fillna(0,inplace=True)
df_clients_ventes_produits[df_clients_ventes_produits['id_prod']=='0_2245']
```

<h3> Création de la colonne Age

```python
df_clients_ventes_produits['age']=2022-df_clients_ventes_produits['birth']
df_clients_ventes_produits
```

<h3> Création de la colonne range d'age

```python
import numpy as np
conditions = [
    (df_clients_ventes_produits['age']>=15) & (df_clients_ventes_produits['age']<=25),
    (df_clients_ventes_produits['age']>25) & (df_clients_ventes_produits['age']<=45),
    (df_clients_ventes_produits['age']>45) & (df_clients_ventes_produits['age']<=65),
    (df_clients_ventes_produits['age']>65)]
values = ['15-25_y', '25-45_y', '45-65_y', '65_y_et_+']
df_clients_ventes_produits["range"]=np.select(conditions,values)
df_clients_ventes_produits
```

<h3> Transformation en datetime

```python
df_clients_ventes_produits.info()
```

```python
import datetime
df_clients_ventes_produits['date']=pd.to_datetime(df_clients_ventes_produits['date'], format = '%Y%m%d %H:%M:%S')
df_clients_ventes_produits
```

<h3> Suppression des heures, minutes et secondes

```python
df_clients_ventes_produits['date']= pd.to_datetime(df_clients_ventes_produits['date'].dt.strftime('%Y-%m-%d'))
df_clients_ventes_produits                                            
```

```python
df_clients_ventes_produits.info()
```

<h3> Création d'une colonne month

```python
df_clients_ventes_produits['month']=df_clients_ventes_produits['date'].dt.month
df_clients_ventes_produits
```

<h3> Création de la colonne month_year

```python
df_clients_ventes_produits['year_month']= df_clients_ventes_produits['date'].dt.strftime('%Y-%m')
df_clients_ventes_produits 
```
